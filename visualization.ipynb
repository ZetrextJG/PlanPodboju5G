```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load the data
df = pd.read_json('output/lambda.json', orient='records', lines=True)
df['population_per_state'] = df['population_sum'] / df['nb_states']
```

```python
#avg by lambda
df_mean = df.groupby('lambda').mean()
df_mean
```

```python
# Plot the data
```

```python
#ax = sns.lineplot(data=df_mean, x='lambda', y='score')
# A more verbose but explicit version of the above not relying on defaults:
ax = sns.lineplot(data=df, x='lambda', y='nb_states', estimator='mean', errorbar=('pi', 50))
ax.set_ylabel('Number of states')
ax.set_title('Number of states by lambda with 50% confidence interval')
ax.set_ylim(0)
ax.figure.autofmt_xdate(rotation=45)
```

```python
#ax = sns.lineplot(data=df_mean, x='lambda', y='score')
# A more verbose but explicit version of the above not relying on defaults:
ax = sns.lineplot(data=df, x='lambda', y='population_sum', estimator='mean', errorbar=('pi', 50))
ax.set_ylabel('Sum of population')
ax.set_title('Summed population by lambda with 50% confidence interval')
ax.set_ylim(0)
ax.figure.autofmt_xdate(rotation=45)
```

```python
#ax = sns.lineplot(data=df_mean, x='lambda', y='score')
# A more verbose but explicit version of the above not relying on defaults:
ax = sns.lineplot(data=df, x='lambda', y='score', estimator='mean', errorbar=('pi', 50))
ax.set_ylabel('Score')
ax.set_title('Score by lambda with 50% confidence interval')
ax.set_ylim(0)
ax.figure.autofmt_xdate(rotation=45)
```

```python
#ax = sns.lineplot(data=df_mean, x='lambda', y='score')
# A more verbose but explicit version of the above not relying on defaults:
ax = sns.lineplot(data=df, x='lambda', y='population_per_state', estimator='mean', errorbar=('pi', 50   ))
ax.set_ylabel('Population per state')
ax.set_title('Population per state  by lambda with 50% confidence interval')
ax.set_ylim(0)
ax.figure.autofmt_xdate(rotation=45)
```

```python
# 3 plots in one row sns
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
sns.lineplot(data=df, x='lambda', y='nb_states', estimator='mean', errorbar=('pi', 50), ax=axs[0])
axs[0].set_ylabel('Number of cities')
axs[0].set_title('Average number of cities with 50% CI')
axs[0].set_ylim(0, 110)
axs[0].figure.autofmt_xdate(rotation=45)

sns.lineplot(data=df, x='lambda', y='population_sum', estimator='mean', errorbar=('pi', 50), ax=axs[1])
axs[1].set_ylabel('Sum of population')
axs[1].set_title('Summed population by lambda with 50% CI')
axs[1].set_ylim(0, 110)
axs[1].figure.autofmt_xdate(rotation=45)

sns.lineplot(data=df, x='lambda', y='population_per_state', estimator='mean', errorbar=('pi', 50), ax=axs[2])
axs[2].set_ylim(0)
axs[2].set_ylabel('Population per city')
axs[2].set_title('Population per city by lambda with 50% CI')


```
